## Classes and objects

```python
class Account:
    bank_balance = 0
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        Account.bank_balance += balance
    def deposit(self, amount):
        # validate
        self.__validate()
        self.balance += amount
        Account.bank_balance += amount
    # Need a new private method validate, which is used inside the Account, should not be used outside
    def withdrawl (self, amount):
        if self.balance < amount:
            print ("Insufficient balance")
        else:
            self.balance -= amount
            Account.bank_balance -= amount
    # Need a new private method validate, which is used inside the Account, should not be used outside
    def __validate(self):
        print ("Validate is called")
    def __str__(self):
        return "Obj of type Account. Name: " + self.name + "; Balance: " + str(self.balance)
    # implement __eq__
    def __eq__(self, other):
        return self.name == other.name

    @staticmethod
    def say_hello():
        print ("Hello")


# Creating the object of class Account
roopesh_ac = Account("roopesh", 1000)
print("Name: ", roopesh_ac.name, "Balance: ", roopesh_ac.balance, "Bank Balance: ", Account.bank_balance)
yash_ac = Account("Yashwant", 2000)
roopesh1_ac = Account("roopesh", 1000)
print("Name: ", yash_ac.name, "Balance: ", yash_ac.balance, "Bank Balance: ", Account.bank_balance)

roopesh_ac.deposit(500)
print("Name: ", roopesh_ac.name, "Balance: ", roopesh_ac.balance)

Account.say_hello()

#Encapsulation
```

## Inheritance & polymmorphism

```python
# we will inherit from account to create Current Account
# Current account has OD Limit

class CurrentAccount(Account):
    def __init__(self, name, balance, od_limit):
        super().__init__(name, balance)
        self.od_limit = od_limit
    def withdrawl(self, amount):
        if self.balance + self.od_limit < amount:
            print ("Insufficient balance")
        else:
            self.balance -= amount

roopesh_cur_ac = CurrentAccount("roopesh", 1000, 5000)
print("Name: ", roopesh_cur_ac.name, "Balance: ", roopesh_cur_ac.balance, "OD Limit: ", roopesh_cur_ac.od_limit)
roopesh_cur_ac.withdrawl(2000)
print("Name: ", roopesh_cur_ac.name, "Balance: ", roopesh_cur_ac.balance, "OD Limit: ", roopesh_cur_ac.od_limit)
```

```python
class ModelLoader:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
            cls._instance.model = None  # Load heavy model once
        return cls._instance

    def load_model(self):
        if self.model is None:
            self.model = "Heavy AI Model Loaded"
        return self.model

# Both variables point to the SAME instance
loader1 = ModelLoader()
print (loader1.model)
loader1.load_model()
print (loader1.model)

loader2 = ModelLoader()
print (loader2.model)
```
